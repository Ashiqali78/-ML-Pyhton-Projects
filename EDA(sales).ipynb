## Exploratory Data Analysis (EDA)

```python
df.columns
```

```python
ex = sns.countplot(x='Gender',data=df)
for bars in ex.containers:
    ex.bar_label(bars)  # bar_label(add lable to the bars)
```

***The graph shows that female purchaes more then the male.***

```python
# step wise 
group_data=df.groupby(['Gender'],as_index=False)
```

```python
sum_data=group_data['Amount'].sum()
```

```python
sales = sum_data.sort_values(by='Amount', ascending=False)
```

```python
sales
```

```python
df.columns
```

```python
# on the basis of age
exg=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in exg.containers:
    exg.bar_label(bars)
```

```python
df.columns
```

```python
# with respect to state
state_sale = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10) 
sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(data=state_sale, x = 'State', y = 'Orders')
```

```python
# Ocupation
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data =df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_sale = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(data=state_sale, x = 'Occupation', y = 'Amount')
```

***So it shows that the most of the buyers are from IT Sector and Then Healthcare and soo on***

```python
## product catagories
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data=df,x='Product_Category')
for lables in ax.containers:
    ax.bar_label(lables)
```

***It shows that the most of the orders are from the Clothing catagories***

```python
state_sale = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.set(rc = {'figure.figsize':(40,10)})
sns.barplot(data=state_sale, x = 'Product_Category', y = 'Amount')
```

***So it shows that the most of the spendings are from food catagories***

```python
state_sale = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data=state_sale, x = 'Product_ID', y = 'Orders')
```
