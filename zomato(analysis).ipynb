```python
# Merge the data from csv and excel files
final_data = pd.merge(df, df1, on = 'Country Code', how = 'left')
```

```python
# Count the total number of occurence of the country
final_data.Country.value_counts()
```

```python
country_val=final_data.Country.value_counts().values
country_val
```

```python
country_names=final_data.Country.value_counts().index
country_names
```

```python
# This will give you the top 3 Countires
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

### observation
***India is the 1st Country where the Zometo got Max Trasections***

```python
final_data.columns
```

```python
# Size is used in groupby 
ratings=final_data.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

### Observation
1. Rattings b/w 4.5 to 5 --> Excellent
2. Rattings b/w 4.0 to 4.4 --> Very Good
3. Rattings b/w 3.5 to 3.9 --> Good
4. Rattings b/w 3.0 to 3.4 --> Average
5. Rattings b/w 2.5 to 2.9 --> Average
6. Rattings b/w 1.8 to 2.4 --> Poor
7. 0 ratings shows that --> No Ratings

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['black','red','orange','yellow','green','green'])
sns.set(rc={'figure.figsize':(15,3)})
```

Observation
1. No rated count is very high
2. The Max number of ratings are b/w 2.5 to 3.4
