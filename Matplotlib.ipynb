```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
# if you want to save this graph
plt.savefig('Scatter.png')
```

*make sure that the size of x and y must be same*

```python
y=x*x
```

```python
plt.plot(x,y,'g-*')
# this 'r--* is the pattern in graph two dashes and one star'
```

```python
# if you want to plot multipul graph in same canves
# 1st graph
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
# 2nd graph
plt.subplot(2,2,2)
plt.plot(x,y,'g*')
# 3rd graph
plt.subplot(2,2,3)
plt.plot(x,y,'yo')
```

```python
# ploting a sin graph
a=np.arange(0,4*np.pi,0.2)
b=np.sin(a)
plt.title('Wave graph')
plt.subplot(2,1,1)
plt.plot(a,b,'r--')
# 2nd 
a=np.arange(0,4*np.pi,0.2)
b=np.cos(a)
plt.subplot(2,1,2)
plt.plot(a,b,'g--')
```

```python
# Bar plot
x = [1,2,3,4]
y = [2,4,6,8]

plt.bar(x,y,color='g')
```

```python
# histogram shows the densityt of the occurance
a=np.array([10,20,14,15,16,11,12,50,65,75,55,56,77,78,100])
plt.hist(a)
plt.title('Histo Gram')
```

```python
# Box Plot 
data=[np.random.normal(0,std,100) for std in range (1,4)]
data
```

```python
plt.boxplot(data)
```

```python
# Pie chart
lables = 'Pyhton','C','Java','Rust'
values = [215,130,225,230]
explor = (0.1,0,0,0)
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
plt.pie(values,labels=lables,colors=colors,autopct='%1.1f%%',explode=explor,shadow=True)
plt.axis('equal')
```
