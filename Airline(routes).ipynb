### Feature Engineering

```python
final_data.head(10)
```

```python
# Spliting the date
final_data['Date']=final_data['Date_of_Journey'].str.split('/').str[0]
final_data['Month']=final_data['Date_of_Journey'].str.split('/').str[1]
final_data['Year']=final_data['Date_of_Journey'].str.split('/').str[2]
```

```python
final_data.head()
```

```python
# Changing the data type of the date
final_data['Date']=final_data['Date'].astype(int)
final_data['Month']=final_data['Month'].astype(int)
final_data['Year']=final_data['Year'].astype(int)
```

```python
# Dropping the data 
final_data.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_data.head()
```

```python
final_data['Arrival_Time'] = final_data['Arrival_Time'].str.split(' ').str[0]
```

```python
# Spliting the time  
final_data['Hour'] = final_data['Arrival_Time'].str.split(':').str[0]
final_data['Minute'] = final_data['Arrival_Time'].str.split(':').str[1]
final_data.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
final_data.head()
```

```python
# Changing the data type of the time
final_data['Hour']=final_data['Hour'].astype(int)
final_data['Minute']=final_data['Minute'].astype(int)
```

```python
final_data['Dep_Houre']=final_data['Dep_Houre'].astype(int)
final_data['Dep_Minute']=final_data['Dep_Minute'].astype(int)
```
